```python
from datasets import load_dataset

# load the commonsense_qa dataset from HuggingFace
ds = load_dataset("tau/commonsense_qa")

# each split is a Dataset object; write them out as json files
for split, dset in ds.items():
    filename = f"commonsense_qa_{split}.json"
    dset.to_json(filename)
    print(f"wrote {len(dset):,} records to {filename}")
```

```python
# load the train split from the json we wrote earlier
train_ds = load_dataset("json", data_files="commonsense_qa_train.json")["train"]

# grab the first 20 examples
train_20 = train_ds.select(range(20))

# write them out to a new file
train_20.to_json("commonsense_qa_train_20.json")
print(f"wrote {len(train_20)} records to commonsense_qa_train_20.json")
```
